### <font color = 'darkgreen'>Selection of the data using `iloc` and `loc`

There are some issue while normal indexing and slicing as shown above which can be avoided if we use iloc and loc for selecting data.

#### iloc: index based selection

Whatever we select using iloc, be it rows or columns, it will be done using index numbers only.

````python
market.head(10)
````

````python
market.iloc[2]
````

````python
market.iloc[2, 4:8]
````

````python
market.iloc[3,7]
````

````python
market.iloc[0:6, 3:8]
````

````python
market.iloc[[1,3,7], [4,6,7,9]]
````

#### loc : label based selection

````python
market.loc[2, ['Profit', 'Sales']]
````

````python
market.loc[10:21, 'Sales': 'Profit']
````

### <font color = 'darkgreen'>Applying Filters on the dataframe

````python
market.head()
````

````python
market['Profit'] < 0
````

````python
#Selecting particular rows that fulfil the criteria
market[market['Profit']<0]
````

````python
(market['Sales'] > 0) & (market['Profit'] < 0)
````

````python
#Applying multiple filters by combining them using 'and' operator
market[(market['Sales'] > 0) & (market['Profit'] < 0)]
````

````python
#Applying multiple filters by combining them using 'or' operator
market[(market['Sales'] > 0) | (market['Profit'] < 0)]
````

````python
#using 'not' operator to do the opposite of the mentioned condition
market[~(market['Profit'] >= 0)]
````

````python
#applying filter on an object column
market[market['Cust_id'] == 'Cust_1818']
````

````python
#Applying the filter on the rows and selecting particular columns for those filtered rows
market.loc[market['Cust_id'] == 'Cust_1818', ['Cust_id', 'Sales', 'Profit', 'Prod_id']]
````

### Concatenating the DataFrames

````python
df2 = pd.DataFrame({'Cust_id': [101, 102, 103, 104, 105],
                   'Profit': [65, 90, 43, 22, 77],
                   'Sales': [987, 889, 345, 776, 587]})
````

````python
df3 = pd.DataFrame({'Cust_id': [106, 107, 108, 109, 110],
                   'Profit': [99, 105, 65, 34, 88],
                   'Sales': [892, 781, 421, 777, 290]})
````

````python
df2
````

````python
df3
````

Here we can see that we have two dataframe which are fulfilling criterias: 
- They have same number of columns.
- Their column names are same.

In such a case it won't be required to merge the data, instead we can simply concatenate them by putting them top and bottom.

````python
pd.concat([df2, df3], axis = 0, ignore_index = True)  #axis = 0 means they will be concatenated top and bottom
````

````python
df4 = pd.DataFrame({'Gender': ['M', 'F', 'F', 'M', 'M'],
                   'Prod_id': ['P1', 'P2', 'P3', 'P4', 'P5']})
````

````python
df2
````

````python
df4
````

If we have to combine df2 and df4, merging won't be required. Instead we can simply put them side by side and concatenate them.

````python
pd.concat([df2, df4], axis = 1) #axis = 1 means concatenation will be done side by side
````

### Grouping and Summarizing the data

````python
merged_data = pd.read_csv('merged_data.csv')
````

````python
merged_data.head()
````

````python
#Get the total profit made by selling all the products
merged_data['Profit'].sum()
````

But what if we want to see the profit made Category-wise. We can group the data based on the required categorical column and then fetch nice and easy to understand summary of the data.

````python
merged_data['Product_Category'].unique()
````

We always group the data based on some categorical column. Here we will use the Product_Category column for grouping.

When we group the data, the data will be split into multiple parts based on the number of unique categories that exist in the selected categorical column.

````python
merged_data['Product_Category'].value_counts()
````

In this case our data 'merged_data' will be split into three groups, 
- OFFICE SUPPLIES with shape (4610, 16)
- TECHNOLOGY with shape (2065, 16)
- FURNITURE with shape (1724, 16)

````python
grp_by_cat = merged_data.groupby('Product_Category')
````

````python
grp_by_cat.groups
````

````python
grp_by_cat.get_group('TECHNOLOGY') #to look at the 'technology' group
````

Now to get the Profit, Category-wise:

````python
grp_by_cat['Profit'].sum()
````

Here grp_by_cat variable contains of all the three groups and the above code goes and selects the `Profit` column from each of the three groups and then sum it up separately.

````python
#to view the above info properly we convert it into a DataFrame
pd.DataFrame(grp_by_cat['Profit'].sum())
````

````python
pd.DataFrame(grp_by_cat['Profit'].sum()).sort_values(by= 'Profit', ascending = False)
````

````python
pd.DataFrame(grp_by_cat['Profit'].sum()).sort_values(by= 'Profit', ascending = False).plot.bar()
````

Can we group the data based on 2 categorical columns?

````python
grp_by_cat_subcat = merged_data.groupby(['Product_Category', 'Product_Sub_Category'])
````

````python
grp_by_cat_subcat['Profit'].sum()
````

````python
pd.DataFrame(grp_by_cat_subcat['Profit'].sum())
````

### Pivot Tables

You may want to use pandas pivot tables as an alternative to groupby(). They provide Excel-like functionalities to create aggregate tables.

````python
help(pd.DataFrame.pivot_table)
````

The general syntax is ```pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', ...)```.
* ```data``` is a dataframe
* ```values``` contains the column to aggregate
* ```index``` is the row in the pivot table
* ```columns``` contains the columns you want in the pivot table
* ```aggfunc``` is the aggregate function

````python
merged_data.pivot_table(values = 'Profit', index = 'Customer_Segment', aggfunc = 'sum')
````

````python
merged_data.pivot_table(values = 'Profit', 
                      index = 'Product_Category', 
                      columns = 'Customer_Segment', 
                      aggfunc = 'sum')
````
